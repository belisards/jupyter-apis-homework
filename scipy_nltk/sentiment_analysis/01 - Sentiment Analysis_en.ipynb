# Classification problem

You have some label and you want to predict it.

# Reading in our data

```python
import pandas as pd
pd.set_option("display.max_colwidth", 200)
pd.set_option("display.max_columns", 200)
```

```python
df = pd.read_csv("data/sentiment140-subset.csv", nrows=100000)
df.shape
```

```python
df.head()
```

# Vectorizing our text data

term frequency inverse document frequency

common words are less important!

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

#vectorizer = TfidfVectorizer(max_features=1000)
# Give me only words that show up at least 20 times
# but are in no more than 70% of the tweets
# these are called "hyperparameters"
# parameters are our features
# hyperparameters are the things we put into the model manually
vectorizer = TfidfVectorizer(min_df=20, max_df=0.7, ngram_range=(1, 3))
vectors = vectorizer.fit_transform(df.text)
words_df = pd.DataFrame(vectors.toarray(), columns=vectorizer.get_feature_names())
# words_df.index = df.text
words_df.head()
```

# Creating and training our model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
```

```python
# Set up X and y
# X => features (in this case, words_df, our word counts)
# y => labels (whether it's positive or negative)
# X is our word counts
# "bag of words" -> "hello how are you" -> ['are', 'hello', 'how', 'you']
X = words_df
# y is a 0/1 for our final label
y = df.polarity
```

```python
# We need to split train and test so that we aren't quizzing it
# on things it's already seen
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
%%time
# clf = LogisticRegression(C=1e9, solver='lbfgs', max_iter=4000)
# clf = RandomForestClassifier(n_estimators=50)
clf = LinearSVC()
# clf = MultinomialNB()
clf.fit(X_train, y_train)
```

# Evaluating our model

```python
# clf.score takes two parameters: features and the right answers
# It's going to take all of our tweets (X)
# and predict whether they're positive or negative, and then compare that
# to the actual positive/negative score(y)

clf.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix

y_true = y_test
y_pred = clf.predict(X_test)
matrix = confusion_matrix(y_true, y_pred)

label_names = pd.Series(['negative', 'positive'])
pd.DataFrame(matrix,
     columns='Predicted ' + label_names,
     index='Is ' + label_names)
```

```python
from sklearn.metrics import confusion_matrix

y_true = y_test
y_pred = clf.predict(X_test)
matrix = confusion_matrix(y_true, y_pred)

label_names = pd.Series(['negative', 'positive'])
pd.DataFrame(matrix,
     columns='Predicted ' + label_names,
     index='Is ' + label_names).div(matrix.sum(axis=1), axis=0)
```

```python
import pandas as pd

df = pd.DataFrame({
    'filename': filenames,
    'content': content
})
df.filename = df.filename.str.replace("corpus-20090418/", "")
# Now we can use
# df.loc['g0pC_taskc.txt'] to grab a specific file
# but we have to use df.index instead of df.filename to get the filename
df = df.set_index('filename')
df.head()
```

## Let's vectorize!!!

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Make a vectorizer
# token_pattern='(?u)\\b\\w+\\b' because we want to match one-character words (a, I, etc)
vectorizer = TfidfVectorizer(ngram_range=(3,6), token_pattern='(?u)\\b\\w+\\b')

# Learn and count the words in df.content
matrix = vectorizer.fit_transform(df.content)
words_df = pd.DataFrame(matrix.toarray(),
                        columns=vectorizer.get_feature_names(),
                        index=df.index)

words_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute the similarities using the word counts
similarities = cosine_similarity(matrix)

# Make a fancy colored dataframe about it
similarity = pd.DataFrame(similarities, index=df.index, columns=df.index)

similarity.style.background_gradient(axis=None, high=0.25)
```

```python
# Give me all of the similarity measures for g0pC_taskc.txt
similarity['g0pE_taske.txt'].sort_values(ascending=False)
```

### STEP ONE: What is the content we want to have as our baseline???

Read it in or pull it from the dataframe, either way. Make sure it's ONE STRING.

```python
content = df.loc['g3pB_taske.txt'].content
content = open("corpus-20090418/g3pB_taske.txt", encoding='latin-1').read()
content[:100]
```

### STEP TWO: Let's learn the words (and n-grams inside) from that ONE document

```python
from sklearn.feature_extraction.text import CountVectorizer

# Find everything between 2-6 words
# and token_pattern means include things like "a" or "I" or "at"
vectorizer = CountVectorizer(ngram_range=(2,6), token_pattern='(?u)\\b\\w+\\b')

# Learn the words
vectorizer.fit([content])
phrases = vectorizer.get_feature_names()
phrases[:10]
```

### STEP THREE: Build a highlighter that knows those phrases

Always have to run this when you read in a new file for your baseline

```python
from rich.console import Console
from rich.highlighter import RegexHighlighter
from rich.theme import Theme

# You can run this code if you have a variable called
# "phrases" that is... the phrases you want to match.
class PhraseHighlighter(RegexHighlighter):
    base_style = "highlighter."
    highlights = [f"(?P<phrase>{phrase})" for phrase in phrases]

theme = Theme({"highlighter.phrase": "on pale_turquoise1"})
highlighter = Console(highlighter=PhraseHighlighter(), theme=theme)
```

### STEP FOUR: Find the documents you are interested in highlighting

Usually this is a subset of the most similar because you don't want to run this on all 2000000000 documents you have

```python
similarity['g0pE_taske.txt'].sort_values(ascending=False).head(20)
```

```python
most_similar = list(similarity['g0pE_taske.txt'].sort_values(ascending=False).head(20).index)
most_similar
```

### STEP FIVE: Run your highlighted on each of those most similar docs

Make sure the index is the filename

```python
for filename in most_similar:
    # the vectorizer doesn't know about , or . or whatever
    # so we'll include those
    content = df.loc[filename].content.lower() \
        .replace(",", "") \
        .replace('"', "") \
        .replace(")", "") \
        .replace("(", "")

    highlighter.print(f"[underline bold]{filename}[/underline bold]")
    highlighter.print(content)
    highlighter.print("-------")
```
